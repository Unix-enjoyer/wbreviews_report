```python
from text2vec import SentenceModel
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Загружаем мультиязычную модель 
model = SentenceModel('shibing624/text2vec-base-multilingual')
```

```python
# похожие предложения (синонимичные)
sent1 = "Мне нужен телефон для установки MAX"
sent2 = "Купить дешевое устройство для мессенджера"
emb1 = model.encode([sent1])[0]
emb2 = model.encode([sent2])[0]
sim = cosine_similarity([emb1], [emb2])[0][0]
print(f"Сходство между похожими предложениями: {sim:.4f}")
```

```python
#разные по смыслу предложения
sent3 = "Модель для предсказывания отзывов."
sent4 = "Вышел месяц из тумана вынул ножик из кармана"
emb3 = model.encode([sent3])[0]
emb4 = model.encode([sent4])[0]
sim2 = cosine_similarity([emb3], [emb4])[0][0]
print(f"Сходство между разными предложениями: {sim2:.4f}")
```

```python
#поиск наиболее близкого кандидата на запрос
query = "Как сменить пароль?"
candidates = [
    "Забыл пароль, что делать?",
    "Как восстановить доступ?",
    "Сегодня солнечно.",
    "Где ближайший супермаркет?"
]
query_vec = model.encode([query])
candidate_vecs = model.encode(candidates)
similarities = cosine_similarity(query_vec, candidate_vecs)[0]
best_idx = np.argmax(similarities)
print(f"Запрос: '{query}'")
print(f"Наиболее похожий кандидат: '{candidates[best_idx]}' (сходство: {similarities[best_idx]:.4f})")
```

```python
# парафраз
paraphrase1 = "Машина не заводится."
paraphrase2 = "Автомобиль отказывается запускаться."
emb_par1 = model.encode([paraphrase1])[0]
emb_par2 = model.encode([paraphrase2])[0]
sim_par = cosine_similarity([emb_par1], [emb_par2])[0][0]
print(f"Сходство парафраза: {sim_par:.4f}")
```

```python
# контрольная пара (совсем разные
unrel1 = "Машина не заводится."
unrel2 = "Кошка сидит на окне."
emb_unrel1 = model.encode([unrel1])[0]
emb_unrel2 = model.encode([unrel2])[0]
sim_unrel = cosine_similarity([emb_unrel1], [emb_unrel2])[0][0]
print(f"Сходство несвязанных предложений: {sim_unrel:.4f}")
```

```python
import numpy as np
import pandas as pd
from sqlalchemy import create_engine
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from tqdm import tqdm
from sentence_transformers import SentenceTransformer
```

```python
# --- Подключение к БД ---
DB_USER = 'admin'
DB_PASSWORD = 'admin123'
DB_HOST = 'localhost'
DB_PORT = '5433'
DB_NAME = 'wb_reviews'

engine = create_engine(f'postgresql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}')
```

```python
# --- Параметры --- не было параметра куда работало 2 минуты с индексами
chunk_size = 10000
total_limit = 10_000
NUM_CATEGORY_OBJECTS = 50

# SQL-запрос (без изменений)
query = f"""
SELECT 
    r.review_text, 
    p.subj_root_name, 
    p.subj_name 
FROM wb_reviews r 
INNER JOIN wb_products p ON r.nm_id = p.nm_id 
WHERE p.subj_name NOT IN ('Цифровой контент', 'Крупная бытовая техника')
GROUP BY r.id, p.subj_name, p.subj_root_name
LIMIT {total_limit}
"""

# --- Загрузка модели sentence-transformers ---
print("Загрузка модели sentence-transformers...")
model_st = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2', device='cuda')
print("Модель загружена.")

```

```python
# --- Контейнеры для данных ---
X_vectors = []   # список эмбеддингов отзывов (float32)
y_root = []      # корневые категории
y_name = []      # подкатегории

print("Чтение данных из БД и векторизация...")
for chunk in pd.read_sql(query, engine, chunksize=chunk_size):
    # Получаем список текстов отзывов из чанка
    texts = chunk['review_text'].tolist()
    
    # Векторизуем батчем (размер батча можно менять, 32 хорошо для скорости/памяти)
    embeddings = model_st.encode(texts, batch_size=32, show_progress_bar=False)
    
    # Преобразуем в float32 и добавляем в общий список
    X_vectors.extend(embeddings.astype(np.float32))
    
    # Сохраняем метки
    y_root.extend(chunk['subj_root_name'].tolist())
    y_name.extend(chunk['subj_name'].tolist())

print("Векторизация завершена.")
```

```python
# --- Преобразование в массивы numpy ---
X = np.array(X_vectors, dtype=np.float32)
y_root = np.array(y_root)
y_name = np.array(y_name)

# Создаём DataFrame для удобства фильтрации
df = pd.DataFrame({
    'X': list(X),          # ссылки на строки массива
    'root': y_root,
    'name': y_name
})
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
print(f"Исходное количество записей: {len(df)}")

# Фильтр 1: корневые категории
root_counts = df['root'].value_counts()
valid_roots = root_counts[root_counts >= NUM_CATEGORY_OBJECTS].index
df = df[df['root'].isin(valid_roots)].reset_index(drop=True)
print(f"После фильтрации по корневым категориям (>= {NUM_CATEGORY_OBJECTS}): {len(df)}")

# Фильтр 2: комбинации root+name
combo_counts = df.groupby(['root', 'name']).size()
valid_combos = combo_counts[combo_counts >= NUM_CATEGORY_OBJECTS].index
df = df[df.set_index(['root', 'name']).index.isin(valid_combos)].reset_index(drop=True)
print(f"После фильтрации по комбинациям (>= {NUM_CATEGORY_OBJECTS}): {len(df)}")

if len(df) == 0:
    raise ValueError("После фильтрации не осталось данных.")

# Создаём отфильтрованные массивы
X_filtered = np.array(df['X'].tolist(), dtype=np.float32)
y_root_filtered = df['root'].values
y_name_filtered = df['name'].values


print(f"Финальное количество записей: {len(X_filtered)}")
print(f"Уникальных корневых категорий: {len(np.unique(y_root_filtered))}")
print(f"Уникальных подкатегорий: {len(np.unique(y_name_filtered))}")
```

```python
# --- Разбиение на train/val/test ---
combined = np.array([f"{r}_{n}" for r, n in zip(y_root_filtered, y_name_filtered)])

X_train, X_temp, y_root_train, y_root_temp, y_name_train, y_name_temp, comb_train, comb_temp = train_test_split(
    X_filtered, y_root_filtered, y_name_filtered, combined,
    test_size=0.3, random_state=42, stratify=combined
)

X_val, X_test, y_root_val, y_root_test, y_name_val, y_name_test, comb_val, comb_test = train_test_split(
    X_temp, y_root_temp, y_name_temp, comb_temp,
    test_size=0.5, random_state=42, stratify=comb_temp
)

# Проверка unseen-подкатегорий
train_names = set(y_name_train)
val_names = set(y_name_val)
test_names = set(y_name_test)

unseen_val = val_names - train_names
unseen_test = test_names - train_names

if unseen_val or unseen_test:
    print(f"Найдены unseen-подкатегории: в val {unseen_val}, в test {unseen_test}")
    mask_val = np.isin(y_name_val, list(train_names))
    X_val = X_val[mask_val]
    y_root_val = y_root_val[mask_val]
    y_name_val = y_name_val[mask_val]
    comb_val = comb_val[mask_val]
    
    mask_test = np.isin(y_name_test, list(train_names))
    X_test = X_test[mask_test]
    y_root_test = y_root_test[mask_test]
    y_name_test = y_name_test[mask_test]
    comb_test = comb_test[mask_test]
    
    print(f"После удаления: val {len(X_val)}, test {len(X_test)}")
    
    if len(X_val) == 0 or len(X_test) == 0:
        raise ValueError("Валидационная или тестовая выборка пуста.")
else:
    print("Все подкатегории в val/test присутствуют в train.")

# --- Кодирование меток ---
le_root = LabelEncoder()
le_name = LabelEncoder()

y_root_train_enc = le_root.fit_transform(y_root_train)
y_root_val_enc = le_root.transform(y_root_val)
y_root_test_enc = le_root.transform(y_root_test)

y_name_train_enc = le_name.fit_transform(y_name_train)
y_name_val_enc = le_name.transform(y_name_val)
y_name_test_enc = le_name.transform(y_name_test)

# --- Обучение моделей ---
clf_root = LogisticRegression(max_iter=1000, multi_class='multinomial', solver='lbfgs')
clf_name = LogisticRegression(max_iter=1000, multi_class='multinomial', solver='lbfgs')

print("Обучение модели для корневой категории...")
clf_root.fit(X_train, y_root_train_enc)

print("Обучение модели для подкатегории...")
clf_name.fit(X_train, y_name_train_enc)

# Оценка на валидации
val_pred_root = clf_root.predict(X_val)
val_pred_name = clf_name.predict(X_val)
acc_root_val = accuracy_score(y_root_val_enc, val_pred_root)
acc_name_val = accuracy_score(y_name_val_enc, val_pred_name)
print(f"Validation accuracy (root): {acc_root_val:.4f}")
print(f"Validation accuracy (name): {acc_name_val:.4f}")

# Финальная оценка на тесте
test_pred_root = clf_root.predict(X_test)
test_pred_name = clf_name.predict(X_test)
acc_root_test = accuracy_score(y_root_test_enc, test_pred_root)
acc_name_test = accuracy_score(y_name_test_enc, test_pred_name)

print(f"\n=== Результаты на тесте ===")
print(f"Точность предсказания корневой категории: {acc_root_test:.4f}")
print(f"Точность предсказания подкатегории: {acc_name_test:.4f}")
```
